````
#Getting the input from the user
a = int(input())
b = int(input())
c = int(input())
n = int(input())
````

````
# Basic For loop
# In this method we use bsic for loop to understand for those who doesnt understand the list comprehension
for i in range(0, a+1):
    for j in range(0, b+1):
        for k in range(0, c+1):
            # if sum of the 3 values is not eqaual to means print i, j, k
            if i+j+k != n:
                print(i, j, k)

````
