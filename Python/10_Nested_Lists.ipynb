* Below code is full executable code for Hacker Rank Python solution

```
final_list = []
listname = []
std = {}
def return_key(val):
    global final_list
    for key, value in std.items():
        if val == value:
            final_list.append(key)

for _ in range(int(input())):
    student_score = []
    name = str(input())
    score = float(input())
    std.update({name:score})

minimum = min(std.values())

for key, value in std.items():
    if minimum == value:
        name = key
        listname.append(name)

for i in listname:
    std.pop(i)
    
sort_std = sorted(std.values())[0]

return_key(sort_std)

for i in sorted(final_list):
    print(i)
```
